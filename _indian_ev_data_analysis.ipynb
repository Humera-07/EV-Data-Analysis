#  2. manefacturer performance

```python
#creat a totla sale columns
evsales["total_sales"] = evsales.iloc[:,2:].sum(axis = 1)
evsales
```

```python
manefecturar_sales = evsales.groupby('Maker')['total_sales'].sum().reset_index()
manefecturar_sales
```

```python
#SORTING MANEFECTURARS BY TOTAL SALES in decending order

manefecturar_sales=manefecturar_sales.sort_values(by = 'total_sales', ascending = False)
manefecturar_sales
```

```python
#bar representation for top ten companies
df = manefecturar_sales.iloc[:10,:]
df.plot(kind='bar', x='Maker', y='total_sales')
plt.xlabel('Category')
plt.ylabel('total sales')
```

#  which company have heighest market shares
